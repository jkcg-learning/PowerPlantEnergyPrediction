```python
from sklearn import preprocessing

# Scale and plot the features against price (target) using the MinMax scaler (Normalization)
min_max_scaler = preprocessing.MinMaxScaler()
col_name = train_data.drop('PE', axis = 1).columns[:]
x = train_data.loc[:, col_name]
y = train_data['PE']

# Normalizing x
x = pd.DataFrame(data = min_max_scaler.fit_transform(x), columns = col_name)

# Examine the normalized data
print(train_data.head())
x.head()
```

```python
fig, axs = plt.subplots(ncols = 2, nrows = 2, figsize = (20, 20))
sns.regplot(y = train_data['PE'], x = x['AT'], ax = axs[0, 0])
sns.regplot(y = train_data['PE'], x = x['V'], ax = axs[0, 1])
sns.regplot(y = train_data['PE'], x = x['AP'], ax = axs[1, 0])
sns.regplot(y = train_data['PE'], x = x['RH'], ax = axs[1, 1])
plt.tight_layout()
```

```python
plt.figure(figsize = (20, 10))
sns.heatmap(train_data.corr(), annot = True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

features = train_data.drop('PE', axis = 1)
target = train_data['PE']
train_features, test_features, train_target, test_target = train_test_split(features, target, test_size = 0.20, random_state = 5)
print("Train Model features shape : ", train_features.shape)
print("Train Model target shape   : ", train_target.shape)
print("Test Model features shape  : ", test_features.shape)
print("Test Model target shape    : ", test_target.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(normalize = True)
model.fit(train_features, train_target)
```

```python
print("Model intercept  : ", model.intercept_, "\n")
print("Model coefficient: ", model.coef_, "\n")

for i in range(len(features.columns)):
    print(features.columns[i], ": ", model.coef_[i])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Model evaluation for training set
train_target_pred = model.predict(train_features)
rmse = (np.sqrt(mean_squared_error(train_target, train_target_pred)))

# Examine the first 10 predicted output from the model
output = pd.DataFrame(train_target[0:10])
output['Predicted'] = train_target_pred[0:10]
output['Difference'] = output['Predicted'] - output['PE']
print(output, "\n")

print("Model training performance:")
print("---------------------------")
print('RMSE is {}'.format(rmse))
print("--------------------------")
```

```python
# Model evaluation for testing set
test_target_pred = model.predict(test_features)
rmse = (np.sqrt(mean_squared_error(test_target, test_target_pred)))

output = pd.DataFrame(test_target[0:10])
output['Predicted'] = test_target_pred[0:10]
output['Difference'] = output['Predicted'] - output['PE']
print(output, "\n")

print("Model testing performance:")
print("--------------------------")
print('RMSE is {}'.format(rmse))
print("--------------------------")
```

```python
from sklearn.linear_model import Ridge
```

```python
model_rr = Ridge(alpha=.001,solver='lsqr',max_iter = 2000,normalize = True) # higher the alpha value, more restriction on the coefficients; low alpha > more generalization, coefficients are barely
# restricted and in this case linear and ridge regression resembles
model_rr.fit(train_features, train_target)
```
